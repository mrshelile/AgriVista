```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

# Create the DataFrame from the list of entries
df = pd.DataFrame(data)

# Filter for 'Cabbage' only and set Date as index
df = df[df['Crop'] == 'Cabbage']
df.set_index('Date', inplace=True)

# Normalize the GPM data
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(df['Gross_Profit_Margin'].values.reshape(-1, 1))

# Prepare the data for LSTM
def create_dataset(data, time_step=1):
    X, y = [], []
    for i in range(len(data) - time_step - 1):
        X.append(data[i:(i + time_step), 0])
        y.append(data[i + time_step, 0])
    return np.array(X), np.array(y)

# Set time step
time_step = 12  # 12 months
X, y = create_dataset(scaled_data, time_step)

# Reshape X for LSTM [samples, time steps, features]
X = X.reshape(X.shape[0], X.shape[1], 1)

# Split into training and test sets
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Build the LSTM model
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(X_train.shape[1], 1)))
model.add(Dropout(0.2))
model.add(LSTM(50, return_sequences=False))
model.add(Dropout(0.2))
model.add(Dense(1))

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(X_train, y_train, epochs=50, batch_size=32)


```

```python
# Make predictions on the test set
predictions = model.predict(X_test)

# Inverse transform the predictions and actual values
predictions = scaler.inverse_transform(predictions)
y_test_original = scaler.inverse_transform(y_test.reshape(-1, 1))

# Calculate error metrics
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

mse = mean_squared_error(y_test_original, predictions)
mae = mean_absolute_error(y_test_original, predictions)
r2 = r2_score(y_test_original, predictions)

print(f'Mean Squared Error: {mse}')
print(f'Mean Absolute Error: {mae}')
print(f'R-squared: {r2}')
```

```python
import pandas as pd
import numpy as np

# Define the crops and districts



# Define the date range
date_range = pd.date_range(start='2022-01-01', end='2024-12-31', freq='MS')

# Create a list to store each entry
data = []

# Base parameters for prices and costs
base_price_range = (10, 30)  # Base market price range
base_cost_range = (3, 10)    # Base production cost range

# Market variability parameters
seasonal_amplitude = 5  # Amplitude of seasonal price variation
trend_slope_price = 0.05  # Gradual increase in prices over time
trend_slope_cost = 0.03   # Gradual increase in costs over time

# Loop through each crop, district, and date to generate the data
for crop in crops:
    for i, date in enumerate(date_range):
            # Calculate seasonal effects
            seasonal_price = seasonal_amplitude * np.sin(2 * np.pi * (date.month / 12))
            seasonal_cost = seasonal_amplitude * np.sin(2 * np.pi * (date.month / 12)) * 0.5
            
            # Calculate market price and production cost with trend
            market_price = (np.random.uniform(*base_price_range) + seasonal_price + 
                            (trend_slope_price * i))
            production_cost = (np.random.uniform(*base_cost_range) + seasonal_cost + 
                               (trend_slope_cost * i))
            
            # Calculate Gross Profit Margin (GPM)
            gpm = (market_price - production_cost) / market_price * 100
            
            entry = {
                'Date': date,
                'Crop': crop,
                'Market_Price': round(market_price, 2),  # Round to 2 decimal places
                'Production_Cost': round(production_cost, 2),  # Round to 2 decimal places
                'Gross_Profit_Margin': round(gpm, 2)  # Round to 2 decimal places
            }
            data.append(entry)

# Create the DataFrame from the list of entries
df = pd.DataFrame(data)

# Display the first few rows of the DataFrame
print(df.head(10))
```
